```python
# Define the bin edges and names
vehicles_bin_edges = [0, 10, 15, 150]
vehicles_bin_names = ['Low Traffic', 'Moderate Traffic', 'High Traffic']

# Set figure background color
sea.set(style="whitegrid", color_codes=True)
sea.set_palette("pastel")

# Use pd.cut() to cut and replace the Vehicles column
dataset['Vehicle Levels'] = pan.cut(dataset.Vehicles, bins=vehicles_bin_edges, labels=vehicles_bin_names)

# Display the first few rows of the modified dataset
dataset.head(5)
```

```python
# Define the mapping
traffic_map = {'Low Traffic': 1, 'Moderate Traffic': 2, 'High Traffic': 3}

# Map the values in the Vehicle Levels column to the new values
dataset['Vehicle Levels'] = dataset['Vehicle Levels'].map(traffic_map)

# Display the first few rows of the modified dataset
dataset.head()
```

```python
dataset['Vehicle Levels'] = dataset['Vehicle Levels'].cat.codes.astype('int64')
```

```python
# importing logistic regression model
from sklearn.linear_model import LogisticRegression
# test train split
from sklearn.model_selection import train_test_split

# Define the features and target variable
X = dataset[['Junction', 'Vehicles', 'ID']]
y = dataset['Vehicle Levels']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Create a logistic regression object
logreg = LogisticRegression()

# Fit the model to the training data
logreg.fit(X_train, y_train)

# Predict the target variable for the testing data
y_pred = logreg.predict(X_test)

# Evaluate the accuracy of the model
accuracy = logreg.score(X_test, y_test)
print("Accuracy:", accuracy)
```
